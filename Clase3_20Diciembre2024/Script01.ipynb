```python
# COnsideremos otro dominio
t = np.linspace(-8, 8 , 250)

# Definamos la variable dependiente
g = t**2 + t**3 -3.14

# Mostremos el diagrama de dispersion
plt.plot(t, g)

plt.show()
```

```python
# En un mismo grafico mostremo g y g+30 para el mismo dominio

plt.plot(t,g, "g",t,g+30,"--r", t , g-30, "--y")
plt.show()
```

```python
# Otro ejemplito

# Definamos otro dominio
dom1 = np.linspace(4,6, 1000)

# Variable dependiente
y1 = np.sin(dom1**2) * np.exp(dom1**(0.3))

#
plt.plot(dom1, y1)
```

```python
# Ejemplito : En una ventana deseo ver (dom1,y1) y en ventana (dom1,derivada_y1)

# Regla de correspondencia de la derivada de y1
derivada_y1 = np.cos(dom1**2)*2*dom1 * np.exp(dom1**(0.3)) + np.exp(dom1**(0.3))*0.3*dom1**(0.7) * np.sin(dom1**2)

# Mostremos las dos ventanas
plt.figure(1)
plt.plot(dom1,y1,"y",
         # Agreguemos una leyenda a la ventana
         label = "Funcion y1")
plt.legend()


plt.figure(2)
plt.plot(dom1,derivada_y1,"g",
         label = "Derivada de y1")

plt.legend()
plt.show()
```

```python
# En una misma ventana mostremos las dos reglas de correspondencia con su etiqueta

# Define la ventana
plt.figure(1)

# Coloquemos la primera regla de correspondencia con algunas caracteristicas propias
plt.plot(dom1, y1,label ="Funcion y1",
         linestyle = "--",
         color = "orange",
        #  marker = "o"
         )

# Coloquemos la segunda regla de correspondencia
plt.plot(dom1,derivada_y1, label = "Derivada de y1",
         linestyle = ":",
         color = "#16620a",
         alpha = 0.3)


# Agreguemos la leyenda
plt.legend()

# Agreguemos algunas etiquetas
plt.xlabel("Variable independiente",{"fontsize":20,"color":"#bd6e65","alpha":0.5})
plt.ylabel("Variable dependiente",{"fontsize":20,"color":"#bd6e65","alpha":0.5})
plt.title("Funciones Matematicas",{"fontsize":24,"color":"#233247"})

# Agreguemos un pequeño rectangulo : Esto se construye usando lineas horizontales
# y lineas verticales
plt.hlines(y= [-20, 20], xmin =4.75, xmax = 5 ,
           linestyles= ":")
plt.vlines(x = [4.75, 5] , ymin = -20, ymax = 20,
           linestyles=":")


# Fin
plt.show()
```

```python
# Ejemplito
# var_random = np.random.uniform(low = -1, high = 10, size = 1000)
var_random = np.random.beta(a = 3.15, b = 9.81, size = 1000)
# Histograma
plt.hist(x = var_random,
        #  COnsideremos que bins sea una variable de tipo cadena de caracter
        # 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', or 'sqrt'.
         bins = "scott",

        #  modifiquemos el color de las barritas
         color = "orange",

        #  Modifiquemos el color del borde de cada patche (rectangulos/barritas)
        edgecolor = "black",

        # Modifiquemos el ancho de la linea que dibuja cada barrita
         linewidth = 2.5
         )


# Fin
plt.show()
```

```python
# Para el siguiente script, fijemos la variable aleatoria
var_random1 = np.random.normal(loc = 2.5, scale = 1.3, size = 1500)
```

```python
# Modifiquemos algunas propiedades de algunas de las barritas que
# compone a un histograma

N, bins, patches = plt.hist(x = var_random1,
         edgecolor = "None",
         rwidth = 0.9)

# Pintemos la tercera barrita de otro color
plt.setp(patches[2], "facecolor","g")

# Pintemos el borde de la primera y ultima barrita de color rojo
plt.setp(patches[0], "edgecolor", "r")
plt.setp(patches[len(N)-1], "edgecolor", "r")

# Agreguemos los extremos de cada intervalo de clase al eje-X : xticks
plt.xticks(bins, rotation= "vertical")

# Fin
plt.show()
```

```python
# Colocar las frecuencias sobre cada barrita que conforma mi histograma

# MOdifiquemos la proporcion de la ventana
plt.figure(figsize = (12,6))

# Construyamos datos
x = np.random.normal(loc = 9.18, scale = 1.8,size = 1000 )
data = ((np.cos(-x**2))/(np.sqrt(2*np.pi*x)))*np.exp(x/2)

# Calculemos los bins usando numpy
bins = np.histogram_bin_edges(data, bins = "scott")

# Mostremos el histograma
_, _, freqs = plt.hist(data, bins = bins, edgecolor = "w", rwidth = 0.8)

# Coloquemos los extreños de los intervalos de clase
plt.xticks(bins, rotation = "vertical", fontsize = 8)

# Coloquemos las frecuencias sobre cada barrita
plt.bar_label(freqs, fontsize = 8, color = "#0acdf4")

# Fin
plt.show()
```
