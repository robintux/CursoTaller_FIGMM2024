```python
# Columna con informacion cualitativa : AIRLINE
flights.AIRLINE.nunique()
```

```python
# El metodo groupby

# Agrupacion : COnsiderar variables de naturaleza cualitativa (AIRLINE)

# Agregar : Considerar variables de naturaleza cuantitativa (ARR_DELAY)

# Funcion de agregacion : mean, max, min,
```

```python
#
airline_arrdelay_mean = flights.groupby("AIRLINE").agg({"ARR_DELAY":"mean"}).sort_values(by = "ARR_DELAY")
airline_arrdelay_mean
```

```python
# tipo de dato
type(airline_arrdelay_mean)
```

```python
airline_arrdelay_mean.plot(kind = "bar")
```

```python
# Agrupamos + Agregamos
flights.groupby("AIRLINE").agg({"ARR_DELAY":"mean"})
```

```python
# Otra forma de obtener este resultado
flights.groupby("AIRLINE").agg({"ARR_DELAY":np.mean})
```

```python
flights
```

```python
# Otra forma
flights.groupby("AIRLINE")["ARR_DELAY"].agg(np.min)
```

```python
flights.groupby("AIRLINE")["ARR_DELAY"].agg(np.sqrt)
```

```python
flights.groupby("AIRLINE")["ARR_DELAY"].agg(np.sum)
```

```python
flights.groupby("AIRLINE")["ARR_DELAY"].agg(np.std)
```

```python
# Juntemos la media y la std por cada grupo/empresa de la columna AIRLINE
# Considerando como variable cuantitativa ARR_DELAY
airline_arrdelay_mean = flights.groupby(["AIRLINE"]).agg({"ARR_DELAY": "mean"})
airline_arrdelay_std = flights.groupby(["AIRLINE"]).agg({"ARR_DELAY": "std"})
airline_arrdelay_var = flights.groupby(["AIRLINE"]).agg({"ARR_DELAY": "var"})
airline_arrdelay_suma = flights.groupby(["AIRLINE"]).agg({"ARR_DELAY": "sum"})
df_mean_std = pd.concat([airline_arrdelay_mean,
                         airline_arrdelay_std,
                         airline_arrdelay_var,
                         airline_arrdelay_suma],axis = 1)
df_mean_std.columns = ["Media", "Desv_Est", "Varianza", "Suma"]
df_mean_std
```

```python
# Deseo encontrar el numero de vuelos cancelados (CANCELLED) para cada aerolinea
# por dia de la semana
flights.groupby(["AIRLINE", "WEEKDAY"]).agg({"CANCELLED":"sum"})
```

```python
# Otra forma
df_1 = flights.groupby(["AIRLINE", "WEEKDAY"])["CANCELLED"].agg("sum")
df_1 = df_1.reset_index()
df_1.loc[df_1.AIRLINE == "F9",:]
```

```python
df_1.loc[df_1.AIRLINE == "F9",:].set_index("WEEKDAY").plot(y = "CANCELLED", kind = "bar")
plt.title("AEROLINEA : F9")
plt.xlabel("Dia de la semana ")
plt.show()
```

```python
# Creacion de un grafico de barras con los vuelos cancelados
# para cada empresa por dia de la semana

for nombre_empresa in flights.AIRLINE.unique():
  df_1.loc[df_1.AIRLINE == nombre_empresa,:].set_index("WEEKDAY").plot(y = "CANCELLED", kind = "bar")
  plt.title("AEROLINEA : "+ nombre_empresa )
  plt.xlabel("Dia de la semana ")
```
