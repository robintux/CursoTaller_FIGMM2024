```python
# IMplementar una funcion que preprocese los pixeles
def preprocessing_pxl(trainX,trainy,testX, testy):
  # Agregamos una dimension al conjunto de variables independientes
  trainX = trainX.reshape(trainX.shape[0],28,28,1 )
  testX = testX.reshape(testX.shape[0], 28,28,1)

  # NOrmalizar los valores
  trainX = trainX.astype("float64")
  testX = testX.astype("float64")
  trainX = trainX/256.0
  testX = testX/256.0

  # One-hot encoding para la variable dependiente
  from keras.utils import to_categorical
  trainy = to_categorical(trainy)
  testy = to_categorical(testy)

  return trainX, trainy, testX, testy
```

```python
# Carguemos algunos modulos
from keras.models import Sequential
from keras.layers import Input, Dense, Conv2D, MaxPooling2D,Flatten
from keras.optimizers import SGD
```

```python
# Implementemos una funcion que permita definir la arquitectura
# y compilar la arquitectura
def define_cnn_base():
  # INstanciamos la clase Sequential : Empaqueta a toda la red neuronal
  model = Sequential()
  # Agregamos la capa de entrada
  model.add(Input(shape = (28,28,1)))
  # Agregamos una capa convolucional + capa de pooling
  model.add(Conv2D(filters=32,
                   kernel_size = (2,2),
                   activation = "relu",
                   kernel_initializer = "he_uniform"
                   ))
  model.add(MaxPooling2D((3,3)))
  # Agregamos una capa convolucional + capa de pooling
  model.add(Conv2D(filters=16,
                   kernel_size = (2,3),
                   activation = "tanh",
                   kernel_initializer = "he_uniform"
                   ))
  model.add(MaxPooling2D((2,2)))
  # Puedo agregar un numero determinado de capas convolucionales + capas pooling

  # Aplana/linealiza las capas provenientes del feature learning
  model.add(Flatten())

  # Definamos un clasificador
  model.add(Dense(units = 50,
                  activation = "elu",
                  kernel_initializer = "ones"))
  model.add(Dense(units = 30,
                  activation = "selu",
                  kernel_initializer = "he_uniform"))
  model.add(Dense(units = 15,
                  activation = "elu",
                  kernel_initializer = "he_uniform"))
  # Agregamos la capa de salida del clasificador
  model.add(Dense(units = 10,
                  activation = "softmax"))

  # Definamos un optimizador en particular , instancia de la clase SGD
  opt_base = SGD(learning_rate= 10**(-3),
                 momentum = 0.01)

  # Compilacion
  model.compile(optimizer = opt_base,
                loss = "categorical_crossentropy",
                metrics = ["accuracy"])

  return model
```

```python
# Carguemos en memoria un objeto para mi CNN
Model1 = define_cnn_base()

# Datos
(trainX, trainy),(testX, testy) = mnist.load_data()
trainX, trainy, testX, testy = preprocessing_pxl(trainX, trainy, testX, testy)

# Ajustamos nuestra primera CNN
Model1.fit(trainX, trainy, epochs = 10, batch_size = 16)
```

```python
y_forecast = Model1.predict(testX)
```

```python
y_forecast[0]
```

```python
# Calculemos un indicador de calidad
from sklearn import metrics

y_forecast_classes = []
for indice in range(len(y_forecast)):
  y_forecast_classes.append([1 if p == max(y_forecast[indice]) else 0 for p in y_forecast[indice]])
y_forecast_classes = np.array(y_forecast_classes)

# Matriz de confucion
metrics.confusion_matrix(testy.argmax(axis =1), y_forecast_classes.argmax(axis = 1))
```
