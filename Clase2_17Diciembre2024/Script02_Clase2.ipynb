````python
# Ejemplito
# Pedir al usuario que ingrese su nota promedio y usando una estructura de decision
# validar si aprobo o desaprobo .
# Aprobado implica que la nota promedio es mayor que 14

# Datos de entrada (input)
nota_promedio = float(input("Ingrese su promedio final : "))

# Logica
if nota_promedio > 14 :
  print("Aprobo")
else :
  print("Desaprobado")
````

````python
# Ejemplito
# Dados los coeficientes de una ecuacion cuadratica : a, b y c
# Considere 3 casos
  # discriminante > 0 : mostrar el mensaje "Raices reales y diferentes"
  # discriminante == 0 : mostrar el mensaje "Raices reales e iguales"
  # discriminante < 0 : mostrar el mensaje "Raices complejas y conjugadas"

# Inputs de mi solucion
a = float(input("Ingresa el coeficiente cuadratico : "))
b = float(input("Ingresa el coeficiente lineal :"))
c = float(input("Ingresa el termino independiente : "))

# Utilicemos la formula para calcular el discriminante
D = b**2 - 4*a*c

# Estructura de decision
if D>0 :
  print("Discriminante = %f \t => Raices reales y diferentes" %(D))
elif D == 0:
  print("Discriminante  = %f \t => Raices reales e iguales" %(D))
elif D < 0 :
  print("Discriminante = %f \t => Raices complejas y conjugadas" %(D))
else:
  print("Caso no contemplado !!!")
````

````python
# Ejemplito
# Dada una variable de tipo cadena de caracter. Escribir todos sus elementos
# en mayusculas

# Definamos una variable de tipo cadena de caracter
cad = "Segunda Clase"

# Barres (usando una estructura de repeticion for) todos los caracteres de la variable cad
# y escribirlos en mayuscula

# Inicializamos nuestro resultado/output como una cadena de caracteres vacia
cad_mayus = ""

# Estructura de repeticion for
for caracter in cad:
  cad_mayus = cad_mayus + caracter.capitalize()

# Mostremos el resultado
print("Mayusculas -> %s" %(cad_mayus))
````

````python
# Ejemplito
# Genere de manera aleatoria (pseudoaleatorios) 20000 numeros enteros en [0,100]
# De como respuesta los numeros multiplos de 17

# Carguemos el modulo randon
import random as rnd

# Almacenemos lo numeros multiplos de 17
num17k = 0

for num in range(20000):
  num_gen = rnd.randint(0,100)
  if num_gen % 17 == 0:
    num17k  = num17k +1

# Mostremos el resultado
print("Numero de multiplos de 17 es  : %d" %(num17k))
````

````python
# Ejemplito  : Aproximacion de pi usando la formula de Leibniz

# Numero de terminos de la serie
num_term = 100000

# Variable acumuladora : PAra almacenar la aproximacion de la constante PI
suma_PI = 0

# Inicializacion el denominador de la serie
k = 1

# Estructura de repeticion para el calculo de la serie
for i in range(num_term):
  if i % 2 == 0:
    suma_PI = suma_PI + 4/k
  else:
    suma_PI = suma_PI - 4/k

  # actualizamos el denominador
  k = k +2

# Mostramos el resultado
print("Aproximacion de Leibniz : %.12f "%(suma_PI))
````
