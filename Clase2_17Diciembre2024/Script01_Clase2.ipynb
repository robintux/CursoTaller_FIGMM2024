```python
# Definamos un objeto de tipo datetime.date que almacene la fecha de hoy para nuestra
# ubicacion geografica
hoy = datetime.date(2024, 12,17)
hoy
```

```python
# Sumemos 60 dias al objeto hoy
dead_line1 = hoy + datetime.timedelta(days = 60)
dead_line1
```

```python
# Restemos 48 dias al objeto hoy
pasado1 = hoy - datetime.timedelta(days = 48)
pasado1
```
