````python
# Simulemos la generacion de 120 notas en [0,20]
Notas2024 = []
for nota in range(120):
  Notas2024.append(rnd.randint(0,20))

Notas2024
````

````python
# Calcular el promedio aritmetico de estas 120 notas
promedio_aritmetico = sum(Notas2024)/len(Notas2024)
promedio_aritmetico
````

````python
# Calculemos el promedio geometrico de estas notas

# Calculemos la productoria de los datos almacenados en Notas2024
prod = 1
for elemento in Notas2024:
  prod = prod * elemento

prod

# Para este conjunto de notas (almacenados en Notas2024) el promedio aritmetico
# tiene el valor de cero
````

````python
# Calculemos el promedio armonica de los datos almacenados en Notas2024

suma_inversas = 1
for elemento in Notas2024:
  suma_inversas =  suma_inversas + elemento**(-1)

media_armonica = (len(Notas2024))/ suma_inversas
media_armonica
````

````python
# Contemos el numero de veces que aparece la nota cero en la variable Notas2024

# Almacenemos el numero de veces que aparece el cero en una variable
num_ceros = 0

for elemento in Notas2024:
  if elemento == 0 :
    num_ceros = num_ceros +1

print("Numero de ceros : %d" %(num_ceros))
````

````python
# Consideremos que la nota aprobatorio es mayor o igual a 14.
# Contemos el numero de aprobados

# Almacenemos el numero de aprobados
num_aprobados = 0

for elemento in Notas2024:
  if elemento >= 14:
    num_aprobados = num_aprobados + 1

print("Numero de aprobados : %d " %(num_aprobados) )
````

````python
# Calculemos el promedio geometrico de los 42 aprobados

prod_aprobados = 1
for elemento in Notas2024:
  if elemento>= 14:
    prod_aprobados = prod_aprobados * elemento

media_geometrica = prod_aprobados**(1/num_aprobados)

print("Media geometrica de los aprobados %.3f" %(media_geometrica))
````

````python
# Mostremos el histograma de los aprobadodos

# Construyamos una lista donde almacenaremos a los aprobados
aprobados = list()

for elemento in Notas2024:
  if elemento >= 14 :
    aprobados.append(elemento)

# Carguemos el modulo matplotlib
import matplotlib.pyplot as plt
plt.hist(aprobados)
````

````python
# Busquemos la manera optima de restar o quitar 1.5 puntos a todos estos aprobados
import numpy as np
np.array(aprobados) - 1.5
````

````python
# Volver al valor original de lista_7
lista_7 = [1,
 'dos',
 'III',
 [4, 5, 6, 7],
 0,
 0.01745189177351341,
 -1.141253236164896,
 datetime.datetime(2024, 12, 24, 0, 27, 33, 521008),
 1,
 'dos',
 'III',
 [4, 5, 6, 7],
 0,
 0.01745189177351341,
 -1.141253236164896,
 datetime.datetime(2024, 12, 24, 0, 27, 33, 521008),
 1,
 'dos',
 'III',
 [4, 5, 6, 7],
 0,
 0.01745189177351341,
 -1.141253236164896,
 datetime.datetime(2024, 12, 24, 0, 27, 33, 521008)]
````

````python
# Eliminemos los datos de tipo str (cadenas de caracteres)

# Primera tarea : Encontrar todos los indices de lista_7 cuyos elementos sean
# datos de tipo str

# Construyamos una lista con los indices de elementos de tipo str
lista_strs = []

# Para acceder a los elementos lo vamos a hacer por intermedio de los indices
for i in range(len(lista_7)):
  # print(type(lista_7[i]))
  if type(lista_7[i]) == str:
    # print(lista_7[i])
    lista_strs.append(i)

print("Lista de indices cuyos elementos son str: ", lista_strs)

````

````python
# Segunda tarea : Con lista_strs  eliminar/removes los elementos de tipo
# str de lista_7
for indice in lista_strs:
  del lista_7[indice]

print("La variable lista_7  sin elementos de tipo str ", lista_7)
````

````python
for elemento in lista_7:
  if type(elemento) == str:
    lista_7.pop(lista_7.index(elemento))

lista_7
````

````python
# Otra forma de atacar el problema es seleccionando los elementos de lista_7 que no sean
# datos de tipo cadena de caracter y almacenarlos en otra lista

lista_7_sin_strs = []

for elemento in lista_7:
  if type(elemento) != str :
    lista_7_sin_strs.append(elemento)

lista_7_sin_strs

````
