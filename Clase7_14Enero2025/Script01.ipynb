# Carguemos datos en memoria

```python
sales_prediction = pd.read_csv("https://raw.githubusercontent.com/robintux/Datasets4StackOverFlowQuestions/refs/heads/master/BIG_MART_SALES_PREDICTION.csv")
sales_prediction.info()
```

```python
# Nombres de columnas
sales_prediction.columns
```

```python
# tipo de dato
type(sales_prediction.columns)
```

```python
sales_prediction.columns[0]
```

```python
lista_nombre_columnas = list(sales_prediction.columns)
lista_nombre_columnas
```

```python
lista_nombre_columnas[0][:4]
```

```python
lista_nombre_columnas[0].split("_")[0]
```

```python
# Separemos en un objeto aquellas columnas que inician con la palabras "Item"

# Crearemos una lista para almacenar los nombres de columna que cumplan con la condicion
lista_Item = []

for nc in lista_nombre_columnas:
  if nc[:4] == "Item":
    lista_Item.append(nc)
lista_Item
```

```python
# Separemos en un dataframe todas las columnas que inicien con la palabra Item
df_Item = sales_prediction.loc[:,lista_Item]
df_Item
```

```python
# Seleccionamos las variables de naturaleza numerica de df_Item
df_Item.info()
```

```python
# EL metodo select_dtypes
df_Item_Num = df_Item.select_dtypes(["float64"])
df_Item_Num
```

```python
# Utilicemos el metodo describe para conocer un poco mas la informacion
# de naturaleza numerica
df_Item_Num.describe()
```

```python
# Calculemos los percentiles 33,66 y 99 del objeto df_Item_Num
df_Item_Num
```

```python
# Veamos los valores faltantes
df_Item_Num.isnull().sum()
```

```python
# Construyamos un diccionario cuyas llaves sean los nombres de columnas de df_Item_Num
dict_data = {}.fromkeys(df_Item_Num.columns)

# Lista para cada fila : tres percentiles
listaq = []
for col in df_Item_Num.columns:
  for q in [0.33,0.66,0.99]:
    listaq.append(np.quantile(df_Item_Num[col],q))
  dict_data[col] = listaq
  listaq = []

dict_data
```

```python
lista_indices = []
for q in [0.33,0.66,0.99]:
  lista_indices.append("q"+ str(q)[-2:])
lista_indices
```

```python
pd.DataFrame(dict_data, index = lista_indices)
```

```python
# Generemos 1000 muestras de tamaño 50
# Y para cada muestra calculemos la media
# Para las columnas FB, TWTR y NFLX

# lista de medias (por muestra) para cada empresa
lista_FB = []
lista_TWTR = []
lista_NFLX = []

for m in range(1000):
  medias = stock_df[["FB", "TWTR" , "NFLX"]].sample(50).mean().to_list()
  lista_FB.append(medias[0])
  lista_TWTR.append(medias[1])
  lista_NFLX.append(medias[2])
```

```python
plt.hist(lista_FB, bins = 100)
plt.show()
```

# API de Matplotlib y Pandas

```python
# Mostremos la dinamica de precios de FB
stock_df.plot(x = "Date",
              y = "FB",
              label = "Precios de Facebook (Res. Diaria)",
              figsize = (15,10),
              color = "#e90bdb"
              )
plt.legend(loc = "upper left")
plt.xlabel("Fecha")
plt.ylabel("Precio [$]")
plt.title("Cotizacion : Empresa META")
plt.grid()
plt.show()
```

```python
# Mostremos los precios de cotizacion  de todas las empresas
stock_df.plot(x = "Date", y = ["FB", "TWTR", "NFLX"],
              figsize = (15,10),
              linewidth = 2
              )
```

```python
# Transformemos la columna Date a un objeto datetime
stock_df.Date = pd.to_datetime(stock_df.Date)
stock_df.info()
```

```python
# Mostremos la informacion de stock_df entre mediados del 2017 y mediados del 2018
import datetime
fecha_interes = stock_df.loc[(stock_df.Date >= datetime.datetime(2017,11,30)) & (stock_df.Date <= datetime.datetime(2018,3,30)), :]

# Mostremos los precios de cotizacion  de todas las empresas
fecha_interes.plot(x = "Date", y = ["FB", "TWTR", "NFLX"],
              figsize = (15,10),
              linewidth = 2
              )
```
