```python
# Generemos 20000 de estos vectores con 10000 notas en [0,20]
# Almacenemos los promedios aritmeticos de estos 20000 vectores

# Guardemos en una lista los 20000 promedios aritmeticos
lista_medias = []

for i in range(20000):
  lista_medias.append(np.random.randint(0,21,10000).mean())

# Calculemos el promedio aritmetico de esos 20000 promedios aritmeticos
np.array(lista_medias).mean()
```

```python
# Calculemos la determinante de una matriz
help(np.linalg.det)
```

```python
# Generemos una matriz de 1000x1000 de numeros aleatorios
mat1 = np.random.randint(0,5, (10**2, 10**2))

# Calculemos la determinante
np.linalg.det(mat1)
```

```python
# Autovalores /Autovectores
help(np.linalg.eigvals)
```

```python
# Calculemos los autovalores de mat1
autovalores = np.linalg.eigvals(mat1)
autovalores.shape
```

```python
# Lista con autovalores reales
eig_real = []
# Separemos los autovalores reales
for autovalor in autovalores:
  if autovalor.imag == 0.0:
    eig_real.append(autovalor)

eig_real

```
