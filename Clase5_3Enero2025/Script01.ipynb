```python
# Almacenemos en una tupla un numero entero n ingresado por el usuario y que cree n
# tuplas dentro de una tupla con 5 elementos aleatorios en [0,666]

# Carguemos el modulo random
import random as rnd

# n = int(input("Ingresa el numero de capas ocultas a generar : "))
n = 3

# Lista para almacenar las tuplas
lista_tuplas = []

for t in range(n):
  t1 = (rnd.randint(0,666),rnd.randint(0,666),rnd.randint(0,666),rnd.randint(0,666),rnd.randint(0,666))
  lista_tuplas.append(t1)

# Empaquetamos la variable lista_tuplas en una tupla
Capas_Ocultas = tuple(lista_tuplas)
Capas_Ocultas
```

# Datos de tipo diccionario

```python
# Definicion : llaves
d1 = {}
```

```python
type(d1)
```

```python
# Caracteristicas
  # Cada elemento de un diccionario esta conformado por un par llave:valor
d2 = {"uno":1, "dos":2, "tres":3}
d2
```

```python
# Acceso a los valores es por intermedio de sus llaves
d2["tres"]
```

```python
# Definamos otro diccionario
datos_personales = {"nombre":"Julio", "edad": 43, "peso":98, "celular":987654321}
```

```python
# Acceso a los valores es por intermedio de sus respectivas llaves
datos_personales["edad"]
```

```python
# Agregar mas informacion
datos_personales["Facultad"] = "FIGMM"
```

```python
datos_personales
```

```python
# Agreguemos el campo "alumnos"
datos_personales["alumnos"] = ["Maria", "Pedro", "Jesus", "William"]
```

```python
datos_personales
```

```python
# Valor asociado a la llave "contrato"
datos_personales["contrato"]
```

```python
# Consideremos una cadena de caracteres como iterable
vocales = "aeiou"
d3 = {}.fromkeys(vocales)
d3
```

```python
d4 = {}.fromkeys(vocales, 666)
d4
```

```python
d5 ={}.fromkeys(vocales, rnd.randint(0,20))
d5
```

```python
# Asignemos un numero aleatorio   a cada valor del diccionario
# que tiene por llaves a vocales
d6 = {}.fromkeys(vocales)

# modifiquemos los valores None del diccionario d6
for llave in vocales:
  d6[llave] = rnd.randint(0,20)

d6
```
