## Grafico de pares ordenados (dispersion)

```python
# Grafico1
# Mostremos un conjunto de pares ordenados
  # Eje-X : Variable independiente (puede ser opcional)
  # Eje-Y : Variable dependiente

# Variable independiente
x = [0.5, 2.3, -1.4]

# Variable dependiente
y = [2.9, 3.6, 9.7]

# Funcion plot del submodulo pyplot
plt.plot(x,y)

# Fin
plt.show()
```

```python
# generemos una lista de colores
lista_colores = ["yellow", "orange", "blue"]

# Para cada color, generaremos una ventana independiente
# con los pares ordenados conformados por x e y
for c in range(len(lista_colores)):
  plt.figure(num = c+1, figsize = ((c+1)*5,(c+1)*5))
  plt.plot(x,y,color = lista_colores[c])
  plt.show()
```

```python
# Grafico2
# Utilizando elementos de numpy

# Dominio de la funcion (conjunto de partida)
# Del intervalo [-5,15] genere 120 puntos equiespaciados
x = np.linspace(-5,15, 120)

# Definamos las imagenes de la funcion
y1 = np.sin(x**(1/2))
y2 = np.exp(x**(0.5))
```

```python
# Dada la naturaleza de las operaciones matematicas realizadas para
# definir y1 e y2, tener en cuenta los mensajes de : "RuntimeWarning: invalid value encountered in sqrt"
```

```python
# Mostremos en una misma ventana (x,y1) y (x,y2)
plt.plot(x,y1,x,y2)
plt.show()
```

```python
# Para agregar una leyenda al grafico debo de utilizar dos funciones plot

plt.plot(x,y1,
         label = "Senoidal",
         color = "red",
         alpha = 0.3)

plt.plot(x,y2,
         label = "Exponencial",
         color = "#e0f508",
         linestyle = "--")

# Mostremos las leyendas
plt.legend()

# Fin
plt.show()
```

## Histogramas

```python
# Funcion hist
plt.hist(x = np.random.normal(loc = 9.81, scale = 2.71, size = 1500))

# Fin
plt.show()
```

```python
# Observar que la ejecucion de la funcion plt.hist devuelve una
# tupla con 3 elementos

# Primer elemento : Numero de elementos de cada intervalo de clase
# Segundo elemento : Limites de cada intervalo de clase
# Tercer elemento : Patches (los rectangulos que contruyen al histograma)
```

```python
# Mostremos otro histograma : Utilicemos otra distribucion de probabilidad
var_random = np.random.gamma(3.15, 1.19, 1500)
N, bins, patches = plt.hist(var_random)
plt.show()
```

```python
# MOdifiquemos el color del primer y ultimo patches (rectangulos)

freq, int_clase , rectangulos = plt.hist(var_random,
                                         bins = "scott",
                                         color = "orange",
                                         edgecolor = "black")

# Agreguemos etiquetas
plt.title("Histograma de una variable aleatorio Gamma")
plt.xlabel("INtervalos de clase")
plt.ylabel("Frecuencias")

# Modificar el color de algunos rectangulos: plt.setp
# Primer rectangulo
plt.setp(rectangulos[0], "facecolor", "g")
# Ultimo rectangulo
plt.setp(rectangulos[len(freq) -1 ], "facecolor", "g")

plt.show()

```

```python
# MOdificar el borde de los rectangulos de mayor valor

freq, int_clase , rectangulos = plt.hist(var_random,
                                         bins = "rice",
                                         color = "orange",
                                         edgecolor = "black",
                                         rwidth = 0.8)

# valor maximo de freq
max_freq = freq.max()
max_freq

# NUmero de veces que el maximo de freq aparece en freq
# Utilicemos una estructura repetitiva para contar el numero de
# veces que aparece el valor maximo
num_veces_max = 0
for elemento in freq:
  if elemento == max_freq:
    num_veces_max = num_veces_max +1
num_veces_max

# Necesito calcular en que indices aparecen esos num_veces_max
# el valor maximo

# Construir una lista de indices donde aparece el valor maximo
lista_ind_max = []

tmp = freq.copy()
for i in range(num_veces_max):
  lista_ind_max.append(tmp.argmax())
  tmp[tmp.argmax()] = tmp.min()
lista_ind_max

# Agreguemos etiquetas
plt.title("Histograma de una variable aleatorio Gamma")
plt.xlabel("INtervalos de clase")
plt.ylabel("Frecuencias")

# Modificar el color de algunos rectangulos: plt.setp
# Primer rectangulo
plt.setp(rectangulos[0], "facecolor", "g")
# Ultimo rectangulo
plt.setp(rectangulos[len(freq) -1 ], "facecolor", "g")

# Con la lista lista_ind_max modifiquemos el color del borde
# de los rectangulos de mayor valor
for j in lista_ind_max:
  plt.setp(rectangulos[j], "edgecolor", "red")
  plt.setp(rectangulos[j], "facecolor", "black")

plt.show()

```
