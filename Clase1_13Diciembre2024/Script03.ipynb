```python
# Mostremos un mensaje de bienvenida
print("Buenas noches. Bienvenidos a la primera sesion .")

# Mostremos los valores de de x e y
print("El valor almacenado en x es : ", x)
print("El valor almacenado en y es : ", y)

# Mostremos una operacion con x  e y
print("El valor almacenado en x es : ", x ,"\nEl valor almacenado en y es :", y, "\nx elevado a la y es : ", x**y)
```

```python
# Controlemos el numero de digitos en la parte decimal
print("El valor almacenado en x es : %.0f \nEl valor almacenado en y es : %.2f \nx elevado a la y es : %.3f"%(x,y, x**y))
```

```python
# Utilicemos docstring para mostrar nuestros resultados
print("""
El valor almacenado en x es : %i
El valor almacenado en y es : %.2f
x elevado a la y es : %.3f
            ====== FIN ======
""" %(x,y,x**y))
```
